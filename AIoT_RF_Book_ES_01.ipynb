```python
# Importa la libreria matematica e quella per il plottaggio
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definisci l'intervallo di valori x come interi
x_values = np.arange(-10, 11)  # Utilizza np.arange() per ottenere valori interi da -10 a 10 inclusi

# Calcola i corrispondenti valori y utilizzando l'equazione della retta
y_values = 7 * x_values - 4

# Crea due array per memorizzare i valori di x e y
x_data = np.array(x_values)
y_data = np.array(y_values)
```

```python
# Disegna il grafico
plt.plot(x_data, y_data, label='y = 7x - 4')

# Aggiungi etichette agli assi
plt.xlabel('x')
plt.ylabel('y')

# Aggiungi una griglia
plt.grid(True)

# Aggiungi una legenda
plt.legend()

# Mostra il grafico
plt.show()
```

```python
# Stampa il contenuto degli array x_data e y_data
print("Contenuto dell'array x_data:")
print(x_data)
print("\nContenuto dell'array y_data:")
print(y_data)
```
