delle classi di CIFAR 10, per l'esercizio, noi utilizzeremo solo due classi: automobili e gatti .
L'ordine delle "label" in Cifar 10 sono: 0 - Airplane (Aereo); 1 - Automobile (Automobile); 2- Bird (Uccello); 3 - Cat (Gatto); 4 - Deer (Cervo); 5 - Dog (Cane); 6 - Frog (Rana); 7 - Horse (Cavallo); 8 - Ship (Nave); 9 - Truck (Camion)

```python
# Importa le librerie utili a visionare il nostro Dataset
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carica il dataset CIFAR-10
(train_ds, test_ds), ds_info = tfds.load('cifar10', split=['train', 'test'], shuffle_files=True, as_supervised=True, with_info=True)

# Definisce una funzione per filtrare le immagini delle automobili
def filter_automobiles(image, label):
    return label == 1  # 1 è la classe per 'automobile' nel dataset CIFAR-10

# Definisce una funzione per filtrare le immagini dei gatti
def filter_cats(image, label):
    return label == 3  # 3 è la classe per 'gatto' nel dataset CIFAR-10

# Filtra il dataset per ottenere solo le immagini delle automobili
train_ds_cars = train_ds.filter(lambda image, label: filter_automobiles(image, label))
test_ds_cars = test_ds.filter(lambda image, label: filter_automobiles(image, label))

# Filtra il dataset per ottenere solo le immagini dei gatti
train_ds_cats = train_ds.filter(lambda image, label: filter_cats(image, label))
test_ds_cats = test_ds.filter(lambda image, label: filter_cats(image, label))
```

```python
# Converte i dataset filtrati in liste per contare le immagini di auto
train_ds_cars = list(train_ds_cars)
test_ds_cars = list(test_ds_cars)

# Converte i dataset filtrati in liste per contare le immagini di gatti
train_ds_cats = list(train_ds_cats)
test_ds_cats = list(test_ds_cats)

# Conta il numero di immagini di automobili nei set di addestramento e di test
num_train_cars = len(train_ds_cars)
num_test_cars = len(test_ds_cars)

# Conta il numero di immagini di gatti nei set di addestramento e di test
num_train_cats = len(train_ds_cats)
num_test_cats = len(test_ds_cats)

# Stampa a schermo il numero di immagini
print(f'Numero di immagini di automobili nel set di addestramento: {num_train_cars}')
print(f'Numero di immagini di automobili nel set di test: {num_test_cars}')

print(f'Numero di immagini di gatti nel set di addestramento: {num_train_cats}')
print(f'Numero di immagini di gatti nel set di test: {num_test_cats}')
```

```python
# Visualizza 40 immagini di gatti del dataset

plt.figure(figsize=(10, 10))  # Crea una nuova figura di dimensioni 10x10 pollici per visualizzare le immagini

# Loop attraverso le prime 40 immagini del dataset `train_ds_cats`
for i, (image, label) in enumerate(train_ds_cats[:40]):
    ax = plt.subplot(5, 8, i + 1)  # Crea una griglia di subplot 5x8 e seleziona il subplot corrente
    plt.imshow(image.numpy())      # Visualizza l'immagine convertendola in un array numpy per compatibilità con matplotlib
    plt.title("Gatto")             # Imposta il titolo del subplot a "Gatto"
    plt.axis("off")                # Rimuove gli assi dal subplot per una visualizzazione più pulita

plt.show()  # Mostra la figura con tutte le immagini visualizzate

# Visualizza 40 immagini di automobili del dataset

plt.figure(figsize=(10, 10))
for i, (image, label) in enumerate(train_ds_cars[:40]):
    ax = plt.subplot(5, 8, i + 1)
    plt.imshow(image.numpy())
    plt.title("Auto")
    plt.axis("off")
plt.show()
```
